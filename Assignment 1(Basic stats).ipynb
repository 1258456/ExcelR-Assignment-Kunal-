# Import Libraries

```python
import pandas as pd 
import numpy as ny
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Dataset

```python
df=pd.read_csv("sales_data_with_discounts.csv")
df
```

## EDA

```python
df.dtypes
```

```python
df.describe()
```

# Data visulization

```python
plt.figure(figsize=(15,5))
plt.tight_layout()
df.hist()
plt.subplots_adjust(hspace=0.5,wspace=0.3)
plt.show()
```

# Histogram

```python
sns.histplot(df["Discount Amount"])
```

# Box plot

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df)
plt.title=("Sales Data With Discount")
```

## Bar Chart

```python
plt.figure(figsize=(10,6))
plt.bar(df["Brand"],df["Avg Price"])
plt.xlabel("Brand")
plt.ylabel("Avg Price")
```

# 

```python
plt.figure(figsize=(10,6))
plt.bar(df["SKU"],df["Volume"])
plt.xlabel("SKU")
plt.ylabel("Volume")
```

# standalization of numerical variable


```python
df.dtypes
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
numerical_cols=['Volume','Avg Price']
Scaler=StandardScaler()
df[numerical_cols]=Scaler.fit_transform(df[numerical_cols])

print("Mean after Standardization:")
print(df[numerical_cols].mean())

print("standard deviation after Standardization:")
print(df[numerical_cols].std())
```

# Dummies


```python
df=pd.get_dummies(df,columns=['Day'],dtype=int)
df
```
